PART B : Q1
Using Python, create your own having columns plant name, sunlight exposure, plant height and answer the following questions:

a.Is there a relationship between the number of hours of sunlight exposure and the height of the plants?



````python
import pandas as pd

# Create a sample dataset
data = {
    'plant_name': ['Fern', 'Cactus', 'Bamboo', 'Rose', 'Tulip', 'Daisy', 'Sunflower', 'Lily', 'Orchid', 'Maple'],
    'sunlight_exposure': [5, 10, 12, 8, 6, 7, 14, 11, 9, 5],  # hours of sunlight
    'plant_height': [30, 150, 200, 60, 50, 40, 180, 70, 40, 20]  # height in cm
}

df = pd.DataFrame(data)

````

Visualize the relationship between sunlight exposure and plant height using a scatterplot.

````python
import matplotlib.pyplot as plt

plt.scatter(df['sunlight_exposure'], df['plant_height'])
plt.title('Sunlight Exposure vs Plant Height')
plt.xlabel('Hours of Sunlight Exposure')
plt.ylabel('Plant Height (cm)')
plt.grid(True)
plt.show()
````

````python
print(df.head())
````

````python
df_reduced = df[['sunlight_exposure', 'plant_height']]
````

````python
df_reduced.corr()
````

````python
correlation = df['sunlight_exposure'].corr(df['plant_height'])
print(f'Correlation coefficient: {correlation}')
print('CORRE',correlation)
````

# Positive Correlation: The coefficient is between 0 and 1.

Negative Correlation: The coefficient is between -1 and 0.

Strong Correlation: Typically considered as |correlation| > 0.7.

Weak Correlation: Typically considered as |correlation| < 0.3.


d.Based on the correlation coefficient, can we conclude that there is a significant association between sunlight exposure and plant growth rate?

````python
if correlation > 0:
    strength = "strong" if abs(correlation) > 0.7 else "weak"
    print(f'There is a {strength} positive correlation between sunlight exposure and plant height.')
elif correlation < 0:
    strength = "strong" if abs(correlation) > 0.7 else "weak"
    print(f'There is a {strength} negative correlation between sunlight exposure and plant height.')
else:
    print('There is no correlation between sunlight exposure and plant height.')
````

# PEARSON CORRELATION for WINE DATASET

````python
import pandas as pd
data2 = pd.read_csv('/content/drive/MyDrive/Dataset/winequality.csv')
````

data2.iloc has two parameters. The first parameter says all rows, and the second parameter requires columns from 1 to 11. The last column has been excluded due to categorical nature of the attribute.

````python
numericdata = data2.iloc[:,0:11]
numericdata.head()
````

````python
datacor = numericdata.corr()
datacor
````

````python
import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(datacor,square=True)
plt.show()
````
